First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site.


```python
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

Above coordinates are just plain numbers that can not give you any intuitive insights about where are those launch sites. If you are very good at geography, you can interpret those numbers directly in your mind. If not, that's fine too. Let's visualize those locations by pinning them on a map.

```python
# Task 2: Mark the success/failed launches for each site on the map
def get_marker_color(outcome):
    if outcome == 1:
        return 'green'
    else:
        return 'red'

for _, launch in spacex_df.iterrows():
    # Coordinates for each launch
    coordinate = [launch['Lat'], launch['Long']]
    
    # Define the color based on success or failure
    marker_color = get_marker_color(launch['class'])
    
    # Add a Circle marker
    folium.Circle(
        location=coordinate,
        radius=1000,
        color=marker_color,
        fill=True,
        fill_color=marker_color
    ).add_to(site_map)
    
    # Add a text label Marker with the launch outcome as a popup
    folium.Marker(
        location=coordinate,
        popup=folium.Popup(f"{launch['Launch Site']}: {launch['class']}", parse_html=True),
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:{marker_color};"><b>{launch["Launch Site"]}</b></div>',
        )
    ).add_to(site_map)

# Display the map
site_map
```

```python
def assign_marker_color(launch_class):
    if launch_class == 1:
        return 'green'
    else:
        return 'red'

# Apply the function to the class column and create a new marker_color column
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
spacex_df.tail()
```

```python
import numpy as np

# Define the Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
import pandas as pd

import pandas as pd
import numpy as np

# Sample DataFrame with launch site names and coordinates
launch_sites_df = pd.DataFrame({
    'Launch Site': ['CCAFS LC-40', 'KSC LC-39A', 'VAFB SLC-4E', 'CCAFS SLC-40'],
    'Lat': [28.5623, 28.5734, 34.6328, 25.9972],
    'Long': [-80.5774, -80.6469, -120.6107, -97.1566]
})

# Create an empty DataFrame to store distances with launch site names as index and columns
site_names_km = launch_sites_df['Launch Site'] + " (km)"
distance_matrix = pd.DataFrame(index=site_names_km, columns=site_names_km)

# Calculate distances between each pair of launch sites
for i, site1 in launch_sites_df.iterrows():
    for j, site2 in launch_sites_df.iterrows():
        if site1['Launch Site'] != site2['Launch Site']:
            distance = calculate_distance(site1['Lat'], site1['Long'], site2['Lat'], site2['Long'])
            distance_matrix.loc[site1['Launch Site'] + " (km)", site2['Launch Site'] + " (km)"] = distance
        else:
            distance_matrix.loc[site1['Launch Site'] + " (km)", site2['Launch Site'] + " (km)"] = 0  # Distance to itself is zero

# Display the distance matrix
distance_matrix


```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon


# Define the coordinates of a launch site and the closest coastline point
launch_site_lat = 28.5623
launch_site_lon = -80.5774
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate the distance
distance_to_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Initialize a map centered around the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# Add a marker for the coastline point with distance information
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_to_coastline:.2f} KM</b></div>',
    )
).add_to(site_map)

# Add a marker for the launch site
launch_site_marker = folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color="blue", icon="rocket", prefix="fa")
).add_to(site_map)

# Display the map
site_map

# Optionally save the map as an HTML file to view in a browser
```
